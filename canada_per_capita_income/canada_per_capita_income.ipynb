## Canada Per Capita Income Prediction
Using Linear Regression to analyze economic growth trends.


### Import Required Libraries



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Load Dataset

```python
data = pd.read_csv("canada_per_capita_income.csv")
data
```

### Dataset Overview

```python
data.info()
data.describe()
data.columns
```

### Feature & Target

```python
X = data[["year"]]
y = data["per capita income (US$)"]
```

### Model Training

```python
model = LinearRegression()
model.fit(X, y)
```

### Model Parameters

```python
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

### Visualization

```python
plt.figure(figsize=(8,5))
plt.scatter(X, y, label="Actual Data")
plt.plot(X, model.predict(X), label="Regression Line")
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Canada Per Capita Income Trend")
plt.legend()
plt.show()
```

### Prediction

```python
future_year = int(input("Enter year: "))
input_df = pd.DataFrame({"year": [future_year]})

predicted_income = model.predict(input_df)
print(f"Predicted Income: {predicted_income[0]:.2f}")
```

## ✅ Conclusion
Linear Regression effectively captures the upward trend in Canada’s per capita income and provides reasonable future estimates.
